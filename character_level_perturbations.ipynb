```python
# import libraries
import nltk
import csv
nltk.download('punkt')  
from random import seed
from random import randint
```

```python
test_file = 'test.txt'
```

## Character Deletion 

```python
# initialise strings
text_1 = ''
deletion_original = ''
# open new files
pertrubed_file = open('character-deletion.txt', 'w')
original_file = open('deletion_original.tsv', 'w')
with open(test_file) as input:
    data = csv.reader(input, delimiter='\t')
    for instance in data:
        topic_sentence = instance[1]

        # tokenize topic sentence
        text_tokenized = nltk.word_tokenize(topic_sentence)
        word_selected = False

        # select random word
        possible_indexes = []
        for i in range(len(text_tokenized)):
            if len(text_tokenized[i]) > 2:
                possible_indexes.append(i)
        index = randint(0, len(possible_indexes)-1)
        chosen_word = text_tokenized[possible_indexes[index]]

        # remove random character
        random_index = randint(1, len(chosen_word)-2)
        new_word_list = chosen_word[:random_index]
        new_word_list += chosen_word[random_index+1:]

        # add elements of list to string
        new_word = "".join(new_word_list)
    
        # reconstruct sentence
        reconstructed_sentence = " ".join(text_tokenized[:possible_indexes[index]])
        
        reconstructed_sentence = f"{reconstructed_sentence} {new_word}" 
        
        after_word = ' '.join(text_tokenized[possible_indexes[index]+1:])
        
        reconstructed_sentence = f"{reconstructed_sentence} {after_word}" 
        
        # output new instance to file
        deletion_original = topic_sentence + '\n'
        text_1 = reconstructed_sentence + '\t'+ instance[2] + '\t' + instance[4] + '\n'
        pertrubed_file.write(text_1)
        original_file.write(deletion_original)

# close file
pertrubed_file.close()
original_file.close()
```

## Character Insertion 

```python
# intialise strings
text_2 = ''
insertion_original = ''
# open new files
perturbed_file = open('character-insertion.txt', 'w')
original_file = open('insertion_original.tsv', 'w')
with open(test_file) as input:
    data = csv.reader(input, delimiter='\t')
    
    for instance in data:
        topic_sentence = instance[1]

        # tokenize topic sentence
        text_tokenized = nltk.word_tokenize(topic_sentence)

        # select random word
        possible_indexes = []
        for i in range(len(text_tokenized)):
            if len(text_tokenized[i]) > 2:
                possible_indexes.append(i)
        index = randint(0, len(possible_indexes)-1)
        chosen_word = text_tokenized[possible_indexes[index]]

        #select random index
        random_index = randint(1, len(chosen_word)-2)
        
        # select random character
        random_char = randint(97, 122)
    
        # add character to word
        new_word_list = chosen_word[:random_index]
        new_word_list += chr(random_char)
        new_word_list += chosen_word[random_index:]
        
        #construct new word
        new_word = "".join(new_word_list)
        
        # reconstruct topic sentence
        reconstructed_sentence = " ".join(text_tokenized[:possible_indexes[index]])
        
        reconstructed_sentence = f"{reconstructed_sentence} {new_word}" 
        
        after_word = ' '.join(text_tokenized[possible_indexes[index]+1:])
        
        reconstructed_sentence = f"{reconstructed_sentence} {after_word}" 
        
        # reconstruct instance
        text_2 = reconstructed_sentence + '\t'+ instance[2] + '\t' + instance[4] + '\n'
        insertion_original = topic_sentence + '\n'
        # add instance to file
        perturbed_file.write(text_2)
        # add original instance to file
        original_file.write(insertion_original)

# close files
perturbed_file.close()
original_file.close()
```

## Claim sentence Character Repetition

```python
# intialise strings
text_3 = ''
repetition_original = ''
#open new files
perturbed_file = open('character-repetition.txt', 'w')
original_file = open('repetition_original.tsv', 'w')

with open(test_file) as input:
    data = csv.reader(input, delimiter='\t')
    for instance in data:
        if (len(instance[2]) > 10):
            
            # tokenize claim sentence
            claim_sentence = instance[2]
            text_tokenized = nltk.word_tokenize(claim_sentence)
            
            # select random word
            possible_indexes = []
            for i in range(len(text_tokenized)):
                if len(text_tokenized[i]) > 2:
                    possible_indexes.append(i)
            index = randint(0, len(possible_indexes)-1)
            if len(possible_indexes) > 0:
                selected_word = text_tokenized[possible_indexes[index]]
            else:
                selected_word = "None applicable"
            
            # select random index from word
            random_index = randint(0, len(selected_word)-1)

            # construct new word
            new_word_list = selected_word[:random_index]
            new_word_list += selected_word[random_index] + selected_word[random_index]
            new_word_list += selected_word[random_index+1:]
            
            #construct new word
            new_word = "".join(new_word_list)

            # reconstruct sentence
            reconstructed_sentence = " ".join(text_tokenized[:possible_indexes[index]])
            
            reconstructed_sentence = f"{reconstructed_sentence} {new_word}" 
            
            after_word = ' '.join(text_tokenized[possible_indexes[index]+1:])
            
            reconstructed_sentence = f"{reconstructed_sentence} {after_word}" 

            # output instance to new file
            text_3 = instance[1] + '\t'+ reconstructed_sentence + '\t' + instance[4] + '\n'
            repetition_original = claim_sentence + '\n'
            perturbed_file.write(text_3)
            original_file.write(repetition_original)

# close file          
perturbed_file.close()
original_file.close()
```

## Letter Case Changing

```python
# intialise strings
output_text_4 = ''
original_lcc = ''
# open new files
perturbed_file = open('letter-case-change.txt', 'w')
original_file = open('lcc_original.tsv', 'w')
with open(test_file) as input:
        
    data = csv.reader(input, delimiter='\t')

    for instance in data:
        if (len(instance[2]) > 11):
            
            # tokenize claim sentence
            claim_sentence = instance[2]
            text_tokenized = nltk.word_tokenize(claim_sentence)

            # select random word
            possible_indexes = []
            for i in range(len(text_tokenized)):
                if len(text_tokenized[i]) > 2:
                    possible_indexes.append(i)
            index = randint(0, len(possible_indexes)-1)
            if len(possible_indexes) > 0:
                chosen_word = text_tokenized[possible_indexes[index]]
            else:
                chosen_word = "None"


            # chose random integer between 1 and 2
            random_change = randint(1, 2)

            new_word_list = ""
            # 1 - only change first character, 2 - change whole word
            if (random_change == 1):
                if (chosen_word[0].islower()):
                    new_word_list = chosen_word[0].upper()
                    new_word_list += chosen_word[1:]
                elif (chosen_word[0].isupper()):
                    new_word_list = chosen_word[0].lower()
                    new_word_list += chosen_word[1:]
                else:
                    new_word_list = chosen_word
            else:
                for i in range(0, len(chosen_word)):
                    if (chosen_word[i].islower()):
                        new_word_list += chosen_word[i].upper()
                    elif (chosen_word[i].isupper()):
                        new_word_list += chosen_word[i].lower()
                    else:
                        new_word_list += chosen_word[i]


            # construct new word
            new_word = "".join(new_word_list)


            # reconstruct sentence
            reconstructed_sentence = " ".join(text_tokenized[:possible_indexes[index]])
            
            reconstructed_sentence = f"{reconstructed_sentence} {new_word}" 
            
            after_word = ' '.join(text_tokenized[possible_indexes[index]+1:])
            
            reconstructed_sentence = f"{reconstructed_sentence} {after_word}" 

            # output instance to file
            output_text_4 = instance[1] + '\t' + reconstructed_sentence + '\t' + instance[4] + '\n'
            original_lcc = claim_sentence + '\n'
            perturbed_file.write(output_text_4)
            original_file.write(original_lcc)
        
# close file      
perturbed_file.close()
original_file.close()
```
